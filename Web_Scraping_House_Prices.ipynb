```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
html = requests.get('https://www.xe.gr/property/results?geo_place_ids%5B%5D=ChIJ8UNwBh-9oRQR3Y1mdkU1Nic&item_type=re_residence&transaction_name=buy').text
```

```python
soup = BeautifulSoup(html,  "html.parser")
houses = soup.find_all('div', class_ = 'common-property-ad')

for house in houses:
    tm = house.find('div', class_='common-property-ad-title').text
    print(tm)
    price = house.find('div', class_='common-property-ad-price').text.replace(" ", '').replace('|','')
    price = price.strip()
    print(price)
    age = house.find('div', class_='grid-x property-ad-construction-year-container')
    age = str(age)
    age = age.replace('div', '')
    age = age.replace('class=', '')
    age = age.replace('<span>','')
    age = age.replace('</span>','')
    age = age.replace('<i "xe xe-house-construction"></i>', '')
    age = age.replace('</>','')
    age = age.replace('×','')
    age = age.replace('< "grid-x property-ad-construction-year-container">','')
    age = age.strip()
    print("Age: " + age)
    location = house.find('span', class_='common-property-ad-address').text.strip()
    location = location.replace(' | Πώληση κατοικίας', '')
    location = location.replace('Αθήνα (', '')
    location = location.replace(')', '')
    print("Location: " + location)
    rooms = house.find('div', class_='grid-x property-ad-bedrooms-container')
    rooms = str(rooms)
    rooms = rooms.replace('<div class="grid-x property-ad-bedrooms-container">', '')
    rooms = rooms.replace('<i class="xe xe-bedroom"></i>', '')
    rooms = rooms.replace('</div>', '')
    rooms = rooms.replace('<span>', '')
    rooms = rooms.replace('</span>', '')
    rooms = rooms.replace('×', '')
    rooms = rooms.strip()
    print("Rooms: " + rooms)
    bathrooms = house.find('div', class_ = 'grid-x property-ad-bathrooms-container')
    bathrooms = str(bathrooms)
    bathrooms = bathrooms.replace('<div class="grid-x property-ad-bathrooms-container">', '')
    bathrooms = bathrooms.replace('<i class="xe xe-bathroom"></i>', '')
    bathrooms = bathrooms.replace('</div>', '')
    bathrooms = bathrooms.replace('<span>', '')
    bathrooms = bathrooms.replace('</span>', '')
    bathrooms = bathrooms.replace('×', '')
    bathrooms = bathrooms.strip()
    print("Bathrooms: " + bathrooms)
    floor = house.find('span', class_='property-ad-level').text.strip().replace(',','')
    print("Floor: " + str(floor) + "\n--------------------------")
```

```python
df = pd.DataFrame({'Area':tm,'Price':price,'Age':age,'Location':location,'Rooms':rooms,'Bathrooms':bathrooms}, index=[0]) 
df.to_csv('houses.csv', index=True, encoding='utf-8')
```
